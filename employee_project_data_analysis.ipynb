```python
# Task 2

# Step 1: Read the Project DataFrame from the CSV
project_df = pd.read_csv("Project.csv")
```

```python
# Step 2: Convert the 'Cost' column to numeric so we can calculate averages
project_df["Cost"] = pd.to_numeric(project_df["Cost"])
```

```python
# Step 3: Loop through the rows to fill missing Cost values using running average
for i in range(len(project_df)):
    # If the current Cost value is NaN (i.e., missing)
    if pd.isna(project_df.loc[i, "Cost"]):
        # Calculate running average from all previous non-missing Cost values
        running_avg = project_df.loc[:i-1, "Cost"].dropna().mean()
        # Fill the missing value with the computed running average
        project_df.loc[i, "Cost"] = running_avg

# Step 4: Display the updated DataFrame with filled Cost values
print(project_df)
```

```python
#Task 3

# Step 1: Read the Employee DataFrame from the CSV
employee_df = pd.read_csv("employee.csv")
```

```python
# Step 2: Split the 'Name' column into two new columns 'FirstName' and 'LastName'
employee_df[['FirstName', 'LastName']] = employee_df['Name'].str.split(' ', n=1, expand=True)

# Step 3: Drop the original 'Name' column
employee_df.drop('Name', axis=1, inplace=True)

# Step 4: Display the final DataFrame
employee_df
```

```python
# Task 4

# Step 1: Load all three CSV files
employee_df = pd.read_csv('Employee.csv')
project_df = pd.read_csv('Project.csv')
seniority_df = pd.read_csv('Seniority.csv')
```

```python
# Step 2: Merge project_df with employee_df on 'ID'
merged_df = pd.merge(project_df, employee_df, on='ID', how='left')
```

```python
# Step 3: Merge the result with seniority_df on 'ID'
Final = pd.merge(merged_df, seniority_df, on='ID', how='left')
```

```python
# Step 4: Display the final combined DataFrame
Final
```

```python
# Task 5

# Step 1: Define a function to calculate bonus
def calculate_bonus(row):
    # Check if Status is 'Finished'
    if row['Status'] == 'Finished':
        # Calculate 5% of the project cost
        return row['Cost'] * 0.05
    else:
        # No bonus for unfinished projects
        return 0

# Step 2: Apply the function to each row of Final DataFrame
Final['Bonus'] = Final.apply(calculate_bonus, axis=1)

# Step 3: Display the updated Final DataFrame
Final
```

```python
# Task 6 

# Step 1: Demote by 1 level where project status is 'Failed'
Final.loc[Final['Status'] == 'Failed', 'Designation Level'] -= 1
```

```python
# Step 2: Drop employees with Designation Level > 4
Final = Final[Final['Designation Level'] <= 4]

# Display Final DataFrame
Final
```

```python
# Task 7

# Step 1: Add prefix based on Gender
Final['FirstName'] = Final.apply(
    lambda row: f"Mr. {row['FirstName']}" if row['Gender'] == 'M' else f"Mrs. {row['FirstName']}",
    axis=1
)
```

```python
# Step 2: Drop the Gender column
Final.drop(columns=['Gender'], inplace=True)

# Step 3: Display updated Final DataFrame
Final
```

```python
# Task 8

# Step 1: Loop through each row using DataFrame index
for i in Final.index:
    # Step 2: Check if Age > 29
    if Final.loc[i, 'Age'] > 29:
        # Step 3: Increase Designation Level by 1
        Final.loc[i, 'Designation Level'] += 1

# Step 4: Display updated Final DataFrame
Final
```

```python
# Task 9 

# Step 1: Group by ID and FirstName, and sum the 'Cost'
TotalProjCost = Final.groupby(['ID', 'FirstName'])['Cost'].sum().reset_index()

# Step 2: Rename the 'Cost' column to 'Total Cost'
TotalProjCost.rename(columns={'Cost': 'Total Cost'}, inplace=True)

# Step 3: Display the result
TotalProjCost
```

```python
# Task 10 

# Step: Filter rows where 'City' contains the letter 'o'
employees_with_o_city = Final[Final['City'].str.contains('o', case=False, na=False)]

# Print the result
employees_with_o_city
```
